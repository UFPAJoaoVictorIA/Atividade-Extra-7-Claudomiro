### Simulação de Chip de Memória: 128M x 4 (512 Mbits)

Esta atividade documenta e simula o funcionamento do chip com base nos seguintes pinos:
- **Endereços:** A0 até A12 (13 pinos multiplexados).
- **Controle de Matriz:** /RAS, /CAS.
- **Seleção de Banco:** Banco 0, Banco 1.
- **Dados:** D0, D1, D2, D3 (Barramento de 4 bits).
- **Controle Geral:** /CS, /WE, /OE.

```python
import math

# Definições baseadas no texto
largura_dados = 4  # D0-D3
profundidade = 128 * 1024 * 1024  # 128M

# Cálculo da capacidade total
capacidade_bits = profundidade * largura_dados
capacidade_mbits = capacidade_bits / (1024 * 1024)

print(f"Capacidade total: {capacidade_mbits} Mbits")
print(f"Organização: {int(profundidade/1024/1024)}M x {largura_dados} bits")
```

```python
class MemoryChip:
    def __init__(self):
        # 13 pinos de endereço (A0-A12)
        self.address_pins = 13
        self.row_address = 0
        self.col_address = 0
        self.bank = 0

        # Sinais de controle
        self.ras_n = True # High (Inativo)
        self.cas_n = True # High (Inativo)
        self.cs_n = True  # Chip Select

    def set_address(self, value):
        self.current_address_bus = value & ((1 << self.address_pins) - 1)

    def pulse_ras(self):
        """Captura o endereço de linha (Row Address Strobe)"""
        self.row_address = self.current_address_bus
        print(f"[RAS] Row Address capturado: {bin(self.row_address)}")

    def pulse_cas(self):
        """Captura o endereço de coluna (Column Address Strobe)"""
        self.col_address = self.current_address_bus
        print(f"[CAS] Column Address capturado: {bin(self.col_address)}")

    def get_full_address(self):
        # Exemplo simplificado de composição de endereço
        return (self.row_address << self.address_pins) | self.col_address

# Demonstração
chip = MemoryChip()
print("--- Simulando acesso à memória ---")
# 1. Coloca endereço de linha no barramento e ativa RAS
chip.set_address(0b1010101010101)
chip.pulse_ras()

# 2. Coloca endereço de coluna no barramento e ativa CAS
chip.set_address(0b0101010101010)
chip.pulse_cas()

print(f"Endereço Interno Final: {hex(chip.get_full_address())}")
```
